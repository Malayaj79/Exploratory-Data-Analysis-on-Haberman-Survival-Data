#Exploratory Data Analysis on Haberman Dataset

---







```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
import warnings 

warnings.filterwarnings("ignore") 
```

```python
#loading dataset
haberman=pd.read_csv('/content/haberman.csv')
print(haberman.head())
```

#Explaination of Important terms in Dataset:

*   **Age**: It represents age of the patient at which they have undergone surgery
*   **Year**: Year in which the patient undergone the surgery
*   **Node**: As lymph fluids leaves the breast it encounters "Nodes" or "Lymph Nodes", these nodes catches any cancer cell present in the fluid and stops it to spread in the body. Having cancer cells present in the nodes represent high risk of cancer spread in the body.
*   **Status**: It represents how much the patient survived after going through the surgery. '1' represnts more than 5 years and '2' represents less than 5 years





```python
long_living_patients=haberman.loc[haberman['status']==1]
short_living_patients=haberman.loc[haberman['status']==2]
```

```python
# Data points and features:
print(haberman.shape)
```

```python
#Counting status of the patients:
print(haberman['status'].value_counts())
```

#Observation:
From the above result we can clearly see that around 81% the patients lived for more than 5 years after the surgery.

```python
print(haberman.columns)
```

```python
#2-D plot
haberman.plot(kind='scatter', x='age',y='nodes')
plt.title('Scatter plot for nodes v/s age')
plt.xlabel('Age')
plt.ylabel('Nodes')
plt.show()
```

```python
#2-D scatterplot with color coding
g=sns.FacetGrid(data=haberman,hue='status').map(plt.scatter,'age','nodes')
plt.legend(['Status: 1','Status: 2'],bbox_to_anchor=(1.5, 1))
plt.title('Scatter plot for nodes v/s age')
plt.xlabel('Age')
plt.ylabel('Nodes')
plt.show()
```

#Observation
The above figure tells us that 'Status' of the patient is not much dependant upon the age of the patient.Hence the diagnosis' effectiveness is not dependent on the age of the patient.

**3-D scatterplot**

```python
sns.set_style('whitegrid')
g=sns.pairplot(haberman,hue='status',palette=['orange','Blue'])
g._legend.remove()
plt.legend(['Status: 1','Status: 2'],bbox_to_anchor=(1.5, 2))
g.fig.suptitle("Pairplot for Haberman dataset", y=1.08)
plt.show()
```

#Mean, variance and Standard Deviation

```python
# Mean nodes:
print("Mean nodes of Status 1 patients:")
print(np.mean(long_living_patients['nodes']))
print("Mean nodes of Status 2 patients:")
print(np.mean(short_living_patients['nodes']))

#Mean age:
print("Mean age of Status 1 patients:")
print(np.mean(long_living_patients['age']))
print("Mean age of Status 2 patients:")
print(np.mean(short_living_patients['age']))

```

#Observation: 
 

1.  As there is a significant difference in the mean of number of nodes present in status 1 and status 2 patients, we can say that lesser the number of nodes higher the chances of the patient surviving longer.
2.   We can observe here that the status of person is independent of the age at which that person got diagnosed.



```python
print("Variance of nodes in Status 1 patients:")
print(np.var(short_living_patients['nodes']))
print("Variance of nodes in Status 2 patients:")
print(np.var(short_living_patients["nodes"]))
```

```python
print("Standard deviation of nodes in Status 1 patients:")
print(np.std(short_living_patients['nodes']))
print("Standard deviation of nodes in Status 2 patients:")
print(np.std(short_living_patients["nodes"]))
```

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(haberman["status"]))
print(np.median(long_living_patients['nodes']))
print(np.median(short_living_patients['nodes']))


print("\nQuantiles:")
print(np.percentile(long_living_patients['nodes'],np.arange(0, 100, 25)))
print(np.percentile(short_living_patients['nodes'],np.arange(0, 100, 25)))


print("\n90th Percentiles:")
print(np.percentile(long_living_patients['nodes'],90))
print(np.percentile(short_living_patients['nodes'],90))


from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(long_living_patients['nodes']))
print(robust.mad(short_living_patients['nodes']))
```

#Observations:


1.   Almost 50% of the Long living patients(Status 1 patients) have 0 nodes
2.   Almost 50% of the Short living patients(Status 2 patients) have less than 4 nodes
